```python
import pandas as pd
# from sklearn.preprocessing import 
```

```python
df = pd.read_csv("./cleaned_data.csv")
df.head()
```

```python
x = df.drop('Customer Type_disloyal Customer', axis='columns')
y = df[['Customer Type_disloyal Customer']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
x_train 
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
lg.fit(x_train, y_train)
```

```python
round(lg.score(x_train,y_train),2)*100  #training set
```

```python
round(lg.score(x_test,y_test),2)*100
```

### OVERFIT PROBLEM
#### TRAINING HIGH SCORE --> 98%
#### TESTING LOW SCORE --> 75%

### UNDERFIT PROBLEM
#### TRAINING LOW SCORE --> 60%

```python
pred = lg.predict(x_test)
pred
```

```python
y_test['prediction'] = pred
```

```python
y_test.head(40)
```

```python
y_test['Customer Type_disloyal Customer'].value_counts()
```

```python
y_test['prediction'].value_counts()
```
