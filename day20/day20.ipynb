```python
cat_col = df.select_dtypes(include="O")  # to get the data who have string
cat_col.head(10)
```

```python
num_col = df.select_dtypes(exclude="O")   # to get the data who have numeric value
num_col.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat_col.head(20)
```

astype(datatype)  = is used to change the datatype of column

```python
cat_col =  pd.get_dummies(cat_col, drop_first="If binary").astype(int)     # convet data into nmeric form 
cat_col.head(20)

```

```python
df2 = pd.concat([cat_col, num_col], axis=1)
df2.head(20)
```

```python
df2['Customer Type_disloyal Customer'].value_counts()
```

```python
disloyal_cust = df2[df2['Customer Type_disloyal Customer'] == 0]
loyal_cust = df2[df2['Customer Type_disloyal Customer'] == 1]
```

```python
loyal_cust.shape   ,   disloyal_cust.shape 
```

```python
disloyal_cust = disloyal_cust.sample(20000)
```

```python
final_df =  pd.concat([loyal_cust,disloyal_cust])
```
